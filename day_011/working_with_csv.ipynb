## 9. Skiprows/nrows Parameter

Problem: You want to skip a specific number of rows from the start or read only a specific number of rows from a large CSV file.

```python
df = pd.read_csv('datasets/aug_train.csv', header=0) # setting 0 as the header

pd.read_csv('datasets/aug_train.csv', skiprows=5) # skips first 5 rows

df
```

```python
pd.read_csv('datasets/aug_train.csv', skiprows=[2, 3, 4]) # skips 2nd, 3rd and 4th rows
```

```python
pd.read_csv('datasets/aug_train.csv', skiprows=lambda x: x % 2 != 0) # skips all odd numbered rows
```

## 13. Handling Dates

Problem: You want to parse date columns while reading a CSV file to ensure they are in the correct datetime format.

```python
pd.read_csv('datasets/IPL Matches 2008-2020.csv').info()
```

we can see the `date` column is of type object (string) instead of datetime.

```python
df = pd.read_csv('datasets/IPL Matches 2008-2020.csv', parse_dates=['date']) # parse the 'date' column as datetime
df.head()
```

```python
df.info()
```

```python
sales = pd.read_csv('datasets/sales.csv').head()
```

```python
# or
sales['date'] = pd.to_datetime(sales[['day', 'month', 'year']])

sales.head()
```

## 14. Convertors 

Problem: You want to apply a custom function to transform the data in a specific column while reading a CSV file.

```python
# dataset before
pd.read_csv('datasets/IPL Matches 2008-2020.csv').head()
```

```python
# function to get initials from name
def rename(name):
    initials = ''.join(word[0].upper() for word in name.split() if word)
    return initials
```

```python
# reading the csv with applying the above function to 'team1', 'team2', 'toss_winner' and 'winner' columns
pd.read_csv('datasets/IPL Matches 2008-2020.csv', converters={'team1': rename, 'team2' : rename, 'toss_winner' : rename, 'winner' : rename})
```
