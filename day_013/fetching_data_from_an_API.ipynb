# **Fetching Data From an API**

```python
# importing libraries
import pandas as pd
import requests
```

```python
# To install python-dotenv package
!pip install python-dotenv
```

```python
# loading environment variable
from dotenv import load_dotenv
import os
load_dotenv()
TMDB_API_KEY = os.getenv('TMDB_API_KEY')
```

```python
# fetching data from the API
response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key={TMDB_API_KEY}&language=en-US&page=1')
```

```python
# json to pandas dataframe
pd.DataFrame(response.json()['results']).head()
```

```python
# empty dataframe
df = pd.DataFrame()
df
```

```python
# fetching data from multiple pages and concatenating to the empty dataframe
for i in range(1,11):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key={TMDB_API_KEY}&language=en-US&page={i}')
    temp_df = pd.DataFrame(response.json()['results'])
    df = pd.concat([df, temp_df], ignore_index=True)
```
